```python
import pandas as pd
import os
#os.chdir('..')
import torch
from torch.utils.data import TensorDataset
from torch.utils.data import DataLoader
import torch.nn as nn

#preprocessing data
from sklearn.preprocessing import StandardScaler

#for plotting
import matplotlib.pyplot as plt
%matplotlib inline
```

## Read in Data

```python
# Read in data 
X = pd.read_csv('data/processed_data/pjm_load_data_2008-11_with_features.csv')
y = pd.read_csv('data/processed_data/load_day.csv')
```

```python
X.head()
```

```python
y.head()
```

## Set up Pytorch model

```python
# Remove dates from X and y
X_copy = X.drop('date', axis = 1)
y_copy = y.drop('date', axis = 1)

# Use last 20% of days as test set, convert to numpy array
index_split = int(0.8*X.shape[0])
X_train = X_copy[:index_split].values
y_train = y_copy[:index_split].values
X_test = X_copy[index_split:].values
y_test = y_copy[index_split:].values
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# Normalize data 
scaler = StandardScaler()
scaler.fit(X_train)
X_train_norm = scaler.transform(X_train)
```

```python
# convert to tensor
X_train_norm = torch.from_numpy(X_train_norm).float()
y_train = torch.from_numpy(y_train).float()
```

```python
## RMS Neural Network 
class RMS_model(nn.Module):
    
    def __init__(self, input_size, hidden_sizes, output_size = 24):
        super().__init__()
        self.layer1 = nn.Linear(input_size, hidden_sizes[0])
        self.layer2 = nn.Linear(hidden_sizes[0], hidden_sizes[1])
        self.layer3 = nn.Linear(hidden_sizes[1], output_size)
    
    def forward(self, x): 
        x = self.layer1(x)
        x = nn.Sigmoid()(x)
        x = self.layer2(x)
        x = nn.ReLU()(x)
        x = self.layer3(x)
        return x
```

```python
torch.manual_seed(2)
# create tensor dataset
td = TensorDataset(X_train_norm, y_train)
dataloader = DataLoader(td, batch_size = X_train_norm.shape[0], shuffle = True) 
```

```python
# initialize
input_size = X_train_norm.shape[1]
hidden_sizes = [20, 20]

model = RMS_model(input_size, hidden_sizes) 
```

```python
learning_rate = 0.001
loss_fn = nn.MSELoss()
# gradient-based optimization method - what takes the gradients
optimizer = torch.optim.Adam(model.parameters(), lr = learning_rate)
```

```python
# initialization
num_epochs = 100
loss_hist = [0] * num_epochs
accuracy_hist = [0] * num_epochs

# train!
for epoch in range(num_epochs):
    for x_batch, y_batch in dataloader:
        # from pytorch page: To use the model, we pass it the input data.
        # This executes the model’s forward, along with some background operations. Do not call model.forward() directly!
        pred = model(x_batch)
        # loss has default reduction mean. That's why we multiply by y_batch_size(0) below.
        loss = loss_fn(pred, y_batch)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        loss_hist[epoch] += loss.item()*y_batch.size(0)
        # calculate % predicted / actual
        is_correct = torch.div(pred,y_batch)
        accuracy_hist[epoch] += is_correct.mean().item()
    loss_hist[epoch] = loss_hist[epoch]**(0.5)/len(dataloader.dataset)
    # fix from dataloader.dataset in book
    accuracy_hist[epoch] /= len(dataloader)
```

```python
#Plot 
fig = plt.figure(figsize = (12, 5))
ax = fig.add_subplot(1, 2, 1)
ax.plot(loss_hist, lw = 3)
ax.set_title("Training loss (RMSE)", size = 15)
ax.set_xlabel('Epoch', size = 15)
ax.tick_params(axis = 'both', which = 'major', labelsize = 15)
ax = fig.add_subplot(1, 2, 2)
ax.plot(accuracy_hist, lw = 3)
ax.set_title("Training Accuracy (Predicted/Actual)", size = 15)
ax.set_xlabel('Epoch', size = 15)
ax.tick_params(axis = 'both', which = 'major', labelsize = 15)
plt.show()
```

## Evaluate the model

```python
# Evaluate the test accuracy 

X_test_norm = scaler.transform(X_test)
X_test_norm = torch.from_numpy(X_test_norm).float()
y_test = torch.from_numpy(y_test).float()

pred_test = model(X_test_norm)
```

```python
# Calculate MSE 
test_RMSE = torch.sqrt(torch.mean(torch.square(pred_test - y_test)))
test_accuracy = torch.mean(torch.div(pred_test, y_test))

print(f'Test RMSE: {test_RMSE} \nTest Accuracy: {test_accuracy}')
```

Overall this already looks pretty good! 

### By Hour of Day

```python
def get_metrics_by_hour(pred, y_actual):
    
    rmse = [0]*24
    accuracy = [0]*24
    
    for hour in range(24): 
        # get predictions and actuals for the hour
        pred_for_hour = pred[:, hour]
        actual_for_hour = y_actual[:, hour]
        rmse[hour] = torch.sqrt(torch.mean(torch.square(pred_for_hour - actual_for_hour))).item()
        accuracy[hour] = torch.mean(torch.div(pred_for_hour, actual_for_hour)).item()
        
    return rmse, accuracy
```

```python
rmse_by_hour, acc_by_hour = get_metrics_by_hour(pred_test, y_test)
```

```python
#Plot 
fig = plt.figure(figsize = (12, 5))
ax = fig.add_subplot(1, 2, 1)
ax.plot(rmse_by_hour, lw = 3)
ax.set_title("Test RMSE", size = 15)
ax.set_xlabel('Hour of Day', size = 15)
ax.tick_params(axis = 'both', which = 'major', labelsize = 15)
ax = fig.add_subplot(1, 2, 2)
ax.plot(acc_by_hour, lw = 3)
ax.set_title("Test Accuracy (Predicted/Actual)", size = 15)
ax.set_xlabel('Hour of Day', size = 15)
ax.tick_params(axis = 'both', which = 'major', labelsize = 15)
plt.show()
```

By hour of day, the RMSE and accuracy is much more noiser and certain hours of day have pretty low accuracy.
